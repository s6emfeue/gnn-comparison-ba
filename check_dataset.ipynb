```python
import torch
import networkx as nx
import matplotlib.pyplot as plt
from torch_geometric.utils import to_networkx
```

```python
dataset = torch.load("DATA/NCI109/processed/NCI109.pt")
```

```python
graph = dataset[0]
```

```python
print(f'Node features (x): {graph.x}')
print(f'Edge indices (edge_index): {graph.edge_index}')
print(f'Edge attributes (edge_attr): {graph.edge_attr if hasattr(graph, "edge_attr") else "None"}')
print(f'Target labels (y): {graph.y}')
print(f'Cycle indices (cycle_index): {graph.cycle_index if hasattr(graph, "cycle_index") else "None"}')
```

```python
data = Data(x=graph['x'], edge_index=graph['edge_index'], edge_attr=graph.get('edge_attr', None), y=graph.get('y', None), cycle_index=graph.get('cycle_index', None))

# Umwandeln des Graphen von PyG in NetworkX
G = to_networkx(data, to_undirected=True)

# Positionen der Knoten festlegen (Kraftbasierte Layout)
pos = nx.spring_layout(G)

# Visualisierung des Graphen
plt.figure(figsize=(10, 10))

# Knoten und Kanten zeichnen
nx.draw(G, pos, with_labels=True, node_size=500, font_size=10, node_color='lightblue', font_weight='bold', edge_color='gray')

# Visualisieren der Knotenmerkmale (optional, je nachdem, was Sie anzeigen möchten)
if hasattr(data, 'x'):
    # Überprüfen, ob 'data.x' ein Tensor ist
    if isinstance(data.x, torch.Tensor):
        node_labels = {i: f'{val.item():.2f}' for i, val in enumerate(data.x.squeeze().tolist())}
    else:
        # Falls 'data.x' eine Liste ist, direkt auf die Werte zugreifen
        node_labels = {i: f'{val:.2f}' for i, val in enumerate(data.x)}
    
    nx.draw_networkx_labels(G, pos, labels=node_labels, font_size=12)

# Visualisieren der Kantenmerkmale (optional, falls vorhanden)
if hasattr(data, 'edge_attr'):
    edge_labels = {i: f'{val.item():.2f}' for i, val in enumerate(data.edge_attr.squeeze().tolist())}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# Visualisierung anzeigen
plt.title("Graph Visualization")
plt.show()
```
